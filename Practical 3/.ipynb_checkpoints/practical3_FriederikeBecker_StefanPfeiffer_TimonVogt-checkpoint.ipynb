# Task 1: Kernel Discriminant Analysis

We will be implementing Kernel Discriminant Analysis, a nonlinear extension to LDA based on the kernel trick, following the original paper by Mika, Rätsch, Weston, Schölkopf and Müller (PDF on StudIP).

The goal of Kernel Discriminant Analysis is find a vector of $\alpha$'s that maximizes
$$ J(\alpha) = \frac{\alpha^T M \alpha}{\alpha^T N \alpha} $$
where 
$$(M_i)_j = \frac{1}{l_i} \sum_{k=1}^{l_i} k(x_j, x_k^i)$$ and
$$N= \sum_{j=1,2} K_j (I - 1_{l_j}) K_j^T$$
Here, the number of samples for class $j$ is denoted by $l_j$. In the definition of $N$, $(K_{j})_{n,m} = k(x_n, x_m^j)$, $I$ is the identity matrix and $1_{l_j}$ is a matrix with each entry equal to $1/l_j$. $M$ has shape $2 \times (l_0 + l_1)$, $N$ has shape $(l_0 + l_1) \times (l_0 + l_1)$.

We obtain the optimal solution by $\alpha = N^{-1}(M_1 - M_0)$. Note that you may have to regularize $N^{-1}$ for numerical stability (see paper or Wikipedia).

1. Implement a function `kda(X, t, kernel)` that carries out Kernel Discriminant Analysis on the data $X,t$ and returns a vector `alpha`.
2. Implement a function `def predict(alpha, x)`
3. Visualize results given a linear kernel and a radial basis function kernel.

Use the following snippet as a starting point.

```python
def kda(X, t, kernel):
    # Set globals for the predict function
    global trainSet
    global activeKernel
    trainSet = X
    activeKernel = kernel
    
    # Start KDA
    l = [(t==0).sum(), (t==1).sum()]
    X_split = X[(t==0)], X[(t==1)]
    M = np.ones((2, len(X)))
    
    # Produce M
    for i in range(len(l)):
        for j in range(len(M[i])):
            M[i][j] = 0
            
            for k in range(l[i]):
                M[i][j] += kernel(X[j], X_split[i][k])
            M[i][j] /= l[i]
    
    
    K = np.ones((len(X), l[0])), np.ones((len(X), l[1]))
    
    # Produce K
    for j in range(len(l)):
        for n in range(len(K[j])):
            for m in range(len(K[j][n])):
                K[j][n][m] = kernel(X[n], X_split[j][m])                
                
    # Produce N
    N = np.zeros((l[0] + l[1], l[0] + l[1]))
    I = (np.identity(l[0]), np.identity(l[1]))
    I_j = (np.identity(l[0]) * (1 / l[0]), np.identity(l[1]) * (1 / l[1]))
        
    for j in range(len(l)):    
        N += K[j].dot(I[j] - I_j[j]).dot(K[j].T)
    
    N_inv = np.linalg.inv(N)
    
    # Produce alpha
    alpha = N_inv.dot((M[1] - M[0]))
    return alpha
    
    

def predict(alpha, X):
    y = []
    X = np.atleast_2d(X)
    for x in X:
        r = 0
        for i in range(len(alpha)):
            r += alpha[i] * activeKernel(trainSet[i], x)
        y.append(r)
    y = np.array(y)
    return (y > 0)
    

def linear_kernel(x_n, x_m):
    return x_n.dot(x_m)

def rbf_kernel(x_n, x_m, gamma=5, **kwargs):
    return np.exp(-np.power(x_n - x_m, 2).sum()/gamma)

alpha = kda(X_train, t_train, linear_kernel)

lin_pred = predict(alpha, X_test)
print("Linear kernel predict: ", np.mean(lin_pred == t_test))

alpha = kda(X_train, t_train, rbf_kernel)

rbf_pred = predict(alpha, X_test)
print("RBF kernel predict: ", np.mean(rbf_pred == t_test))

# Visualize
fig, ax = plt.subplots(1,2, figsize=(17, 7))

# Points
ax[0].scatter(X_test[:, 0], X_test[:, 1], c=lin_pred)
ax[0].set_title("Linear kernel predict")

# Points
ax[1].scatter(X_test[:, 0], X_test[:, 1], c=rbf_pred)
ax[1].set_title("RBF kernel predict")

alpha = 0
```

# Task 2: Support Vector Machines

Next, you will implement a Support Vector Machine (SVM). For solving the quadratic program, we provide a simple interface to the cvxopt library below.

In SVMs, each data sample $x_n$ has a corresponding parameter $\alpha_n$ which indicates if $x_n$ is a support vector. In the latter case $\alpha_n > 0$ holds. 
The goal of learning the SVM is to figure out which samples are support vectors by learning $\mathbf{\alpha}$. The dual SVM optimizes the following quadratic program.

$$ \min \frac{1}{2} \sum_{n=1}^N \sum_{m=1}^N \alpha_n \alpha_m t_n t_m k(\mathbf{x}_n, \mathbf{x}_m) - \sum_{n=1}^N \alpha_n$$
subject to 
$$ 0 \leq \alpha_n \leq C $$
$$ \sum_{n=1}^N \alpha_n t_n = 0 $$ 

The quadratic program solver expects the following form:
$$ \min \frac{1}{2} \alpha^T P \alpha - \mathbf q^T \mathbf \alpha $$
subject to 

$$A \alpha = b$$
$$G \alpha \leq h $$

Here, $A$ and $G$ are matrices with one row per individual constraint. Similarly, $b$ and $h$ are vectors with one element per individual constraint.


#### Task 2.1
 
Use the code provided below as a basis to express the constrained optimization problem in terms of $P, q, A, b, G$ and $h$ and implement a function `fit_svm` which passes these variables to the provided QP solver.

**Hints:**
  - The box constraint $0 \leq \alpha_n \leq C$ defines two constraints of the form $G \alpha_n \leq h$ for each $\alpha_n$.
  - The inequality $x \geq 0$ is equivalent to $-x \leq 0$.

```python
def linear_kernel(a, b):
    return a.dot(b)

def rbf_kernel(a, b, gamma=5, **kwargs):
    return np.exp(-np.power(a - b, 2).sum()/gamma)

def fit_svm(X, t, kernel, C=1.0):
    '''Fit SVM using data (X,t), specified kernel and parameter C.'''

    t = np.array([-1 if l == 0 else 1 for l in t])
    n_samples = len(X)
    
    P = np.array([[t[n] * t[m] * kernel(X[n], X[m]) for m in range(len(X))] for n in range(len(X))])
    q = -1 * np.ones((len(X), 1))
    A = np.array([t], dtype='float')
    b = np.array([[0]], dtype='float')
    G = np.concatenate((-1 * np.identity(len(X)), np.identity(len(X))))
    h = np.array([0 for k in range(len(X))] + [C for k in range(len(X))])
    

    assert P.shape == (len(X), len(X))
    assert len(q) == len(X)
    assert A.shape == (1, n_samples) and A.dtype == 'float'
    assert b.shape == (1, 1)
    assert len(G) == 2 * len(X)
    assert len(h) == 2 * len(X)

    return solve_quadratic_program(P, q, A, b, G, h)

def solve_quadratic_program(P, q, A, b, G, h):
    '''Uses cvxopt to solve the quadratic program.'''
    P, q, A, b, G, h = [cvxopt.matrix(var) for var in [P, q, A, b, G, h]]
    minimization = cvxopt.solvers.qp(P, q, G, h, A, b)
    lagr_mult = np.ravel(minimization['x'])
    print("")
    return lagr_mult


def extract_parameters(X, t, kernel, lagr_mult):
    '''Computes the intercept from the support vector constraints.
    
    Inputs
        X:         predictors
        t:         targets
        kernel:    a kernel to be used
        lagr_mult: the Lagrange multipliers obtained by solving the dual QP
    
    Returns
        lagr_mult: (same as input) -> only those of support vectors
        svs:       set of support vectors
        sv_labels: targets t_n for the support vectors
        intercept: computed intercept
    '''
    t = np.array([-1 if l == 0 else 1 for l in t])
    
    # "Define" zero
    zero = 1e-5
    
    # Support vectors
    sv = lagr_mult > zero
    a = lagr_mult[sv]
    svs = X[sv]
    sv_labels = t[sv]
    
    # Intercept
    b = 0
    for n in range(len(a)):
        #b += sv_labels[n]
        #b -= np.sum(a * sv_labels * kernel(svs, svs[n]))
        tmp = sv_labels[n]
        for m in range(len(a)):
            tmp -= a[m] * sv_labels[m] * kernel(svs[m], svs[n])
        b += tmp        
        
    b /= len(a)
    intercept = b
    lagr_mult = a

    return lagr_mult, svs, sv_labels, intercept


# Fit the SVM
lm = fit_svm(X_train, t_train, rbf_kernel)
params = extract_parameters(X_train, t_train, rbf_kernel, lm)
```

#### Task 2.2

Having learnt an SVM, we can use the calculated parameters to make predictions on novel samples.
- Implement a function `svm_predict(X, kernel, lagr_mult, svs, sv_labels, intercept)`.
- Use this function to compute the accuracy of a linear SVM on the test set.

```python
def svm_predict(X, kernel, lagr_mult, svs, sv_labels, intercept):
    y = []
    for x in X:
        tmp = intercept
        for s in range(len(svs)):
            tmp += lagr_mult[s] * sv_labels[s] * kernel(svs[s], x)
        y.append(tmp)
        #y.append(intercept + np.sum(lagr_mult * sv_labels * kernel(svs, x)))
    y = np.array(y)
    return (y > 0)

X_train, X_test, t_train, t_test = split_data(X_2d, t_2d, seed=1)

lm = fit_svm(X_train, t_train, linear_kernel)
params = extract_parameters(X_train, t_train, linear_kernel, lm)

pred_test = svm_predict(X_test, linear_kernel, *params)
plt.scatter(X_test[:,0], X_test[:, 1], c=pred_test)
print('score linear kernel', np.mean(pred_test == t_test))
```

# Task 3: Decision Trees

Next, we will implement a simple decision tree classifier using the Wine dataset, one of the standard sklearn datasets. 

We will use the Gini impurity as a criterion for splitting. It is defined for a set of labels as
$$ G = \sum_{i=0}^C p(i) * (1- p(i)) $$

Given labels $l$ and split $l_a$ and $l_b$, the removed impurity can be computed by $l - l_a + l_b$

Here is a simple explanation of the Gini impurity that you may find useful: https://victorzhou.com/blog/gini-impurity/


#### Task 3.1

1. Plot the distribution of the first feature of for each class of the wine dataset.
2. Implement a function `gini_impurity(t)` that computes the Gini impurity for an array of labels `t`.
3. Calculate the removed Gini impurity for a split after 50 samples.

```python
# Load Wine dataset and split into train+test set
from sklearn.datasets import load_wine
X, t = load_wine(return_X_y=True)
X_train, X_test, t_train, t_test = split_data(X, t)

plt.hist(X[: ,0], bins=50)
plt.title("Distribution of the first feature")
plt.show()

# Note: number of labels: 3 (0.0, 1.0, 2.0)
def gini_impurity(t):
    if(len(t) == 0):
        return 0.0
        
    p = len(t[t == 0]) / len(t), len(t[t == 1]) / len(t), len(t[t == 2]) / len(t)
    return p[0] * (1 - p[0]) + p[1] * (1 - p[1]) + p[2] * (1 - p[2])
    
# Split after 50 samples
t_1 = t[0:50]
t_2 = t[51:]

print("Removed Gini impurity for a split after 50 samples: ", (gini_impurity(t) - gini_impurity(t_1) - gini_impurity(t_2)))
```

#### Task 3.2
Plot the removed Gini impurity for the first 12 features. Which is the optimal split?

```python
# Using weighted Gini impurities
def removed_gini_after(A, idx):
    if(len(A) == 0):
        return 0.0
    
    A_1 = A[idx]
    A_2 = A[~idx]
    l = len(A)
    l_1 = len(A_1)
    l_2 = len(A_2)
    return gini_impurity(A) - ((l_1 / l) * gini_impurity(A_1)) - ((l_2 / l) * gini_impurity(A_2))


fig, ax = plt.subplots(3, 4, figsize=(12,8))
fig.tight_layout(h_pad = 2)

for n_feat in range(12):
    this_ax = ax[n_feat % 3, n_feat % 4]
    this_ax.set_ylim(0, 0.3)
    
    x = np.linspace(0, 20, 100)
    y = [removed_gini_after(t_train, (X_train[:, n_feat] > i)) for i in x]
    
    this_ax.plot(x, y)
    this_ax.set_title("Optimal: " + str(round(x[np.argmax(y)], 3)))
```
